# YUV

```python
img = cv2.imread('img/t.png', 3)
rgbimg = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
yuv_img = cv2.cvtColor(img, cv2.COLOR_BGR2YUV)

fig = plt.figure(figsize=(5, 5))
plt.title('YUV')
m2 = plt.imshow(yuv_img)
plt.show()


plt.subplot(1,3,1)
plt.title('Y channel')
imgy = yuv_img[:, :, 0]
#m2 = plt.imshow(yuv_img[:, :, 0])
m2 = plt.imshow(imgy)
plt.subplot(1,3,2)
plt.title('U channel')
m2 = plt.imshow(yuv_img[:, :, 1])
plt.subplot(1,3,3)
plt.title('V channel')
m2 = plt.imshow(yuv_img[:, :, 2])
plt.show()
```

```python
hsv_img = cv2.cvtColor(rgbimg, cv2.COLOR_BGR2HSV)

fig = plt.figure(figsize=(5, 5))
m2 = plt.imshow(hsv_img)
plt.show()
```
