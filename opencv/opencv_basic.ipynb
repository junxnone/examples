# Rotate

```python
def rotate_img(img, rotate_angle):
  rows, cols = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY).shape
  M = cv2.getRotationMatrix2D((cols / 2, rows / 2), rotate_angle, 1)
  dst = cv2.warpAffine(img, M, (cols, rows))
  return dst

fig = plt.figure(figsize=(10, 10))

img = cv2.imread('img/t.png', 3)
plt.subplot(1,3,1)
m1 = plt.imshow(img)

dst = rotate_img(img,90)
plt.subplot(1,3,2)
m1 = plt.imshow(dst)
plt.show()

cv2.imwrite("./img/t90.png",dst)
```

# 使用numpy的方式旋转90度避免长宽

```python
img = cv2.imread('img/t.png', 3)

img90=np.rot90(img)
img180=np.rot90(img90)
img270=np.rot90(img180)


fig = plt.figure(figsize=(15, 15))

plt.subplot(1,3,1)
m1 = plt.imshow(img90)

plt.subplot(1,3,2)
m1 = plt.imshow(img180)

plt.subplot(1,3,3)
m1 = plt.imshow(img270)

plt.show()
```
